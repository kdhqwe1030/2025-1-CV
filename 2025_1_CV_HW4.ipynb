```python
import os
import time
import torch
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import Image, display
import cv2
import numpy as np


# 1. 환경 설정 및 YOLOv5 설치
# 1. 환경 설정 및 YOLOv5 설치
# 1. 환경 설정 및 YOLOv5 설치
# 1. 환경 설정 및 YOLOv5 설치
# 1. 환경 설정 및 YOLOv5 설치


def setup_environment():
    """
    YOLOv5 환경 설정 및 설치

    작동 원리:
    1. YOLOv5 GitHub 저장소를 클론
    2. 필요한 패키지들을 설치
    3. GPU 사용 가능 여부 확인
    """
    print("=" * 60)
    print("YOLOv5 환경 설정 시작")
    print("=" * 60)

    # YOLOv5 저장소 클론 (이미 존재하면 넘어감)
    if not os.path.exists('yolov5'):
        print("YOLOv5 저장소 클론 중...")
        os.system('git clone https://github.com/ultralytics/yolov5')
    else:
        print("YOLOv5 저장소가 이미 존재합니다.")

    # 작업 디렉토리 변경
    os.chdir('yolov5')
    print(f"현재 작업 디렉토리: {os.getcwd()}")

    # 필요 패키지 설치
    print("필요한 패키지 설치 중...")
    os.system('pip install -r requirements.txt')

    # GPU 사용 가능 여부 확인
    device = 'cuda' if torch.cuda.is_available() else 'cpu'
    print(f"사용 가능한 디바이스: {device}")

    if device == 'cuda':
        print(f"GPU 메모리: {torch.cuda.get_device_properties(0).total_memory / 1024**3:.1f}GB")

    return device


# 2. 데이터셋 준비
# 2. 데이터셋 준비
# 2. 데이터셋 준비
# 2. 데이터셋 준비
# 2. 데이터셋 준비

def prepare_dataset():
    """
    COCO128 데이터셋 다운로드 및 준비

    작동 원리:
    1. COCO128 데이터셋을 다운로드 (COCO 데이터셋의 작은 버전)
    2. 압축 해제
    3. 데이터셋 구조 확인

    COCO128 데이터셋:
    - 128개의 이미지로 구성된 COCO 데이터셋의 샘플
    - 80개 클래스 (사람, 자동차, 동물 등)
    - 빠른 실험과 테스트에 적합
    """
    print("\n" + "=" * 60)
    print("데이터셋 준비 시작")
    print("=" * 60)

    # COCO128 데이터셋 다운로드
    if not os.path.exists('coco128'):
        print("COCO128 데이터셋 다운로드 중...")
        os.system('curl -L "https://github.com/ultralytics/yolov5/releases/download/v1.0/coco128.zip" -o coco128.zip')
        os.system('unzip -q coco128.zip')
        print("데이터셋 다운로드 완료")
    else:
        print("COCO128 데이터셋이 이미 존재합니다.")

    # 데이터셋 구조 확인
    print("\n데이터셋 구조:")
    print(f"   - 학습 이미지 수: {len(os.listdir('coco128/images/train2017'))}")
    print(f"   - 라벨 파일 수: {len(os.listdir('coco128/labels/train2017'))}")

    # 클래스 정보 확인
    with open('data/coco128.yaml', 'r') as f:
        content = f.read()
        print(f"   - 클래스 수: 80개 (사람, 자전거, 자동차, 오토바이, 비행기 등)")


# 3. 모델 학습
# 3. 모델 학습
# 3. 모델 학습
# 3. 모델 학습
# 3. 모델 학습


def train_model(epochs=50, batch_size=16, img_size=640):
    """
    YOLOv5 모델 학습

    매개변수:
    - epochs: 학습 에포크 수
    - batch_size: 배치 크기
    - img_size: 입력 이미지 크기

    작동 원리:
    1. 사전 훈련된 YOLOv5s 모델을 기반으로 전이 학습
    2. COCO128 데이터셋으로 파인튜닝
    3. 학습 중 성능 지표 자동 계산 및 저장
    4. 최고 성능 모델 자동 저장
    """
    print("\n" + "=" * 60)
    print("모델 학습 시작")
    print("=" * 60)

    # 학습 파라미터 출력
    print(f"학습 설정:")
    print(f"   - 모델: YOLOv5s")
    print(f"   - 에포크: {epochs}")
    print(f"   - 배치 크기: {batch_size}")
    print(f"   - 이미지 크기: {img_size}")
    print(f"   - 데이터셋: COCO128")

    # 학습 시작 시간 기록
    start_time = time.time()

    # 학습 명령어 실행
    train_command = f"""python train.py \
    --img {img_size} \
    --batch {batch_size} \
    --epochs {epochs} \
    --data coco128.yaml \
    --weights yolov5s.pt \
    --project ./runs/train \
    --name my_yolo_exp \
    --save-period 10"""

    print(f"학습 명령어 실행: {train_command}")
    os.system(train_command)

    # 학습 시간 계산
    training_time = time.time() - start_time
    print(f"총 학습 시간: {training_time:.2f}초 ({training_time/60:.1f}분)")

    return training_time


# 4. 성능 평가
# 4. 성능 평가
# 4. 성능 평가
# 4. 성능 평가
# 4. 성능 평가

def evaluate_model():
    """
    학습된 모델의 성능 평가

    작동 원리:
    1. 학습 결과 파일들을 읽어와서 성능 지표 추출
    2. mAP, Precision, Recall 등의 지표 분석
    3. 학습 곡선 시각화
    """
    print("\n" + "=" * 60)
    print("모델 성능 평가")
    print("=" * 60)

    # 결과 파일 경로
    results_path = 'runs/train/my_yolo_exp'

    # 학습 결과 CSV 파일 읽기
    try:
        results_df = pd.read_csv(f'{results_path}/results.csv')
        results_df.columns = results_df.columns.str.strip()  # 공백 제거

        # 최종 성능 지표 출력
        final_metrics = results_df.iloc[-1]

        print("최종 성능 지표:")
        print(f"   - mAP@0.5: {final_metrics['metrics/mAP_0.5']:.4f}")
        print(f"   - mAP@0.5:0.95: {final_metrics['metrics/mAP_0.5:0.95']:.4f}")
        print(f"   - Precision: {final_metrics['metrics/precision']:.4f}")
        print(f"   - Recall: {final_metrics['metrics/recall']:.4f}")

        return final_metrics

    except Exception as e:
        print(f"결과 파일 읽기 오류: {e}")
        return None

def visualize_training_results():
    """
    학습 결과 시각화

    작동 원리:
    1. YOLOv5가 자동 생성한 그래프들을 표시
    2. 학습 곡선, PR 곡선, 혼동 행렬 등을 보여줌
    """
    print("\n" + "=" * 60)
    print("학습 결과 시각화")
    print("=" * 60)

    results_path = 'runs/train/my_yolo_exp'

    # 학습 곡선은 따로 크게 표시
    results_img = f'{results_path}/results.png'
    if os.path.exists(results_img):
        print("학습 곡선 (Loss, mAP, Precision, Recall):")
        display(Image(results_img))

    # 나머지 3개 이미지를 한 행에 배치
    fig, axes = plt.subplots(1, 3, figsize=(18, 6))

    side_by_side_images = [
        ('confusion_matrix.png', '혼동 행렬'),
        ('PR_curve.png', 'Precision-Recall 곡선'),
        ('F1_curve.png', 'F1 점수 곡선')
    ]

    for i, (img_name, title) in enumerate(side_by_side_images):
        img_path = f'{results_path}/{img_name}'
        if os.path.exists(img_path):
            img = plt.imread(img_path)
            axes[i].imshow(img)
            axes[i].set_title(title)
            axes[i].axis('off')
        else:
            axes[i].text(0.5, 0.5, f'{img_name} 파일을 찾을 수 없습니다.',
                        ha='center', va='center')
            axes[i].set_title(title)
            axes[i].axis('off')

    plt.tight_layout()
    plt.show()


# 5. 추론 및 속도 측정
# 5. 추론 및 속도 측정
# 5. 추론 및 속도 측정
# 5. 추론 및 속도 측정
# 5. 추론 및 속도 측정

def test_inference_speed():
    """
    추론 속도 측정

    작동 원리:
    1. 테스트 이미지들에 대해 추론 실행
    2. 처리 시간을 측정하여 FPS 계산
    3. 검출 결과 이미지 생성
    """
    print("\n" + "=" * 60)
    print("추론 속도 측정")
    print("=" * 60)

    # 테스트 이미지로 추론 실행
    detect_command = """python detect.py \
    --weights runs/train/my_yolo_exp/weights/best.pt \
    --img 640 \
    --source coco128/images/train2017 \
    --project ./runs/detect \
    --name test_results \
    --save-txt \
    --save-conf"""

    print("추론 실행 중...")
    start_time = time.time()
    os.system(detect_command)
    inference_time = time.time() - start_time

    # 처리된 이미지 수 계산
    test_images = len(os.listdir('coco128/images/train2017'))
    fps = test_images / inference_time

    print(f"추론 성능:")
    print(f"   - 총 처리 시간: {inference_time:.2f}초")
    print(f"   - 처리 이미지 수: {test_images}장")
    print(f"   - FPS: {fps:.2f}")
    print(f"   - 이미지당 평균 처리 시간: {inference_time/test_images*1000:.1f}ms")

    return fps, inference_time

def show_detection_results():
    """
    검출 결과 이미지 표시

    작동 원리:
    1. 추론 결과로 생성된 이미지들을 표시
    2. 바운딩 박스와 클래스 레이블이 그려진 결과 확인
    """
    print("\n" + "=" * 60)
    print("검출 결과 이미지")
    print("=" * 60)

    detect_path = 'runs/detect/test_results'

    if os.path.exists(detect_path):
        result_images = os.listdir(detect_path)[:5]  # 처음 5개 이미지만 표시

        for img_name in result_images:
            if img_name.endswith(('.jpg', '.png')):
                print(f"검출 결과: {img_name}")
                img_path = f'{detect_path}/{img_name}'
                display(Image(img_path))
    else:
        print("검출 결과 폴더를 찾을 수 없습니다.")


# 6. 성능 요약 출력
# 6. 성능 요약 출력
# 6. 성능 요약 출력
# 6. 성능 요약 출력
# 6. 성능 요약 출력

def print_performance_summary(training_time, final_metrics, fps, inference_time):
    """
    성능 요약 출력

    작동 원리:
    1. 모든 실험 결과를 종합
    2. 표 형태로 정리하여 출력
    3. 과제 제출용 요약 데이터 제공
    """
    print("\n" + "=" * 60)
    print("성능 요약")
    print("=" * 60)

    # 모델 정보
    model_info = {
        "모델": "YOLOv5s",
        "데이터셋": "COCO128",
        "클래스 수": "80개",
        "학습 이미지 수": "128장"
    }

    print("실험 설정:")
    for key, value in model_info.items():
        print(f"   - {key}: {value}")

    # 성능 지표
    if final_metrics is not None:
        print(f"\n성능 지표:")
        print(f"   - mAP@0.5: {final_metrics['metrics/mAP_0.5']:.4f}")
        print(f"   - mAP@0.5:0.95: {final_metrics['metrics/mAP_0.5:0.95']:.4f}")
        print(f"   - Precision: {final_metrics['metrics/precision']:.4f}")
        print(f"   - Recall: {final_metrics['metrics/recall']:.4f}")

    # 속도 성능
    print(f"\n속도 성능:")
    print(f"   - 학습 시간: {training_time/60:.1f}분")
    print(f"   - 추론 속도: {fps:.2f} FPS")
    print(f"   - 이미지당 처리 시간: {inference_time/128*1000:.1f}ms")

    # 모델 크기 정보
    model_path = 'runs/train/my_yolo_exp/weights/best.pt'
    if os.path.exists(model_path):
        model_size = os.path.getsize(model_path) / (1024*1024)  # MB
        print(f"   - 모델 크기: {model_size:.1f}MB")

    print(f"\n저장된 파일들:")
    print(f"   - 학습된 모델: runs/train/my_yolo_exp/weights/best.pt")
    print(f"   - 학습 결과 그래프: runs/train/my_yolo_exp/results.png")
    print(f"   - 검출 결과 이미지: runs/detect/test_results/")
    print("\n" + "=" * 60)
    print("성능 요약 ")
    print("=" * 60)

    # 모델 정보
    model_info = {
        "모델": "YOLOv5s",
        "데이터셋": "COCO128",
        "클래스 수": "80개",
        "학습 이미지 수": "128장"
    }

    print(" 실험 설정:")
    for key, value in model_info.items():
        print(f"   - {key}: {value}")

    # 성능 지표
    if final_metrics is not None:
        print(f"\n 성능 지표:")
        print(f"   - mAP@0.5: {final_metrics['metrics/mAP_0.5']:.4f}")
        print(f"   - mAP@0.5:0.95: {final_metrics['metrics/mAP_0.5:0.95']:.4f}")
        print(f"   - Precision: {final_metrics['metrics/precision']:.4f}")
        print(f"   - Recall: {final_metrics['metrics/recall']:.4f}")

    # 속도 성능
    print(f"\n 속도 성능:")
    print(f"   - 학습 시간: {training_time/60:.1f}분")
    print(f"   - 추론 속도: {fps:.2f} FPS")
    print(f"   - 이미지당 처리 시간: {inference_time/128*1000:.1f}ms")

    # 모델 크기 정보
    model_path = 'runs/train/my_yolo_exp/weights/best.pt'
    if os.path.exists(model_path):
        model_size = os.path.getsize(model_path) / (1024*1024)  # MB
        print(f"   - 모델 크기: {model_size:.1f}MB")

    print(f"\n 저장된 파일들:")
    print(f"   - 학습된 모델: runs/train/my_yolo_exp/weights/best.pt")
    print(f"   - 학습 결과 그래프: runs/train/my_yolo_exp/results.png")
    print(f"   - 검출 결과 이미지: runs/detect/test_results/")


# 7. 메인 실행 함수
# 7. 메인 실행 함수
# 7. 메인 실행 함수
# 7. 메인 실행 함수
# 7. 메인 실행 함수

def main():
    """
    전체 실험 파이프라인 실행

    실행 순서:
    1. 환경 설정
    2. 데이터셋 준비
    3. 모델 학습
    4. 성능 평가
    5. 결과 시각화
    6. 추론 속도 측정
    7. 최종 요약 출력
    """
    print("YOLOv5 객체탐지 모델 학습 실험 시작")

    try:

        device = setup_environment()# 1. 환경 설정
        prepare_dataset() # 2. 데이터셋 준비
        training_time = train_model(epochs=50, batch_size=16, img_size=640) # 3. 모델 학습 (파라미터는 필요에 따라 조정 가능)
        final_metrics = evaluate_model() # 4. 성능 평가
        visualize_training_results() # 5. 학습 결과 시각화
        fps, inference_time = test_inference_speed() # 6. 추론 속도 측정
        show_detection_results() # 7. 검출 결과 표시
        print_performance_summary(training_time, final_metrics, fps, inference_time) # 8. 최종 요약 출력


    except Exception as e: print(f"실험 중 오류 발생: {e}")

# 실행
if __name__ == "__main__":
    main()
```
